### CROSSTABLE () --> PANDAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset("titanic")
dataset.head()
```

```python
dataset.shape
```

```python
datos = pd.crosstab(dataset["sex"], dataset["alive"])
datos
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
sns.barplot(x=datos["no"].index, y=datos["no"].values ,hue=datos["no"].index ,legend=True , ax=axes[0])
axes[0].set_xlabel("Sexo")
axes[0].set_ylabel("Cantidad")
axes[0].set_title("Gráfico 1: Datos 'no'")
sns.barplot(x=datos["yes"].index, y=datos["yes"].values ,hue=datos["yes"].index ,legend=True , ax=axes[1])
axes[1].set_xlabel("Sexo")
axes[1].set_ylabel("Cantidad")
axes[1].set_title("Gráfico 2: Datos 'yes'")
plt.show()
```
